# Exercise 2: The table contains a column called 'categories' and each entry in this column is populated by a list. We are interested in those businesses that restaurants. Create a new column 'Restaurant_type' that contains a description of the restaurant based on the other elements of 'categories. 
## That is, if we have '[Sushi Bars, Japanese, Restaurants]' in categories the 'Restaurant_type will be '{'SushiBars': 1, 'Japanese': 1, 'Mexican': 0, ...}'

```python
import numpy as np
df['Restaurant_type']=np.nan
df['Restaurant_type'] = df['Restaurant_type'].astype(dict)
df.head(100)
```

```python
import operator
from collections import defaultdict
```

```python
Rest_types=[]

for row in df['categories']:
    if row:
        if 'Restaurants' in row:
            for item in row:
                if item!='Restaurants':
                    if item not in Rest_types:
                        Rest_types.append(item)


for i,row in enumerate(df['categories']):
    dict_rest=defaultdict(int)
    if row:
        if 'Restaurants' in row:
            for rest_typ in Rest_types:
                if rest_typ in row:
                    dict_rest[rest_typ]=1
                else:
                    dict_rest[rest_typ]=0
            dict_rest=sorted(dict_rest.items(), key=operator.itemgetter(1), reverse=True)
            df.set_value(i, 'Restaurant_type', dict_rest)
            


```

# Exercise 3: Lets clean the 'attributes' column. The entries in this column are dictionaries. We need to do two things: 
## 1) Turn all the True or False in the dictionary to 1s and 0s.
## 2) There are some entries within dictionaries that are dictionaries themselves, lets turn the whole entry into just one dictionary, for example if we have 
### '{'Accepts Credit Cards': True, 'Alcohol': 'none','Ambience': {'casual': False,'classy': False}}' 
### then turn it into
### '{'Accepts Credit Cards':1, 'Alcohol_none': 1, 'Ambience_casual': 0, 'Ambience_classy': 0}'. 
### There might be other entries like {'Price Range': 1} where the values are numerical so we might want to change that into {'Price_Range_1': 1}.

*The reason we modify categorical variables like this is that machine learning algorithms cannot interpret textual data like "True" and "False". They need numerical inputs such as 1 and 0.*

```python
df['attributes_new']=np.nan


    
```

```python
df['attributes_new'] = df['attributes_new'].astype(dict)
```

```python
#df['new_attributes']=transform_T_F(df[attributes])


import ast 

def key_value(key, value):
    if value.strip()=='True':
        return key.strip(), 1
    elif value.strip()=='False':
        return key.strip(), 0
    else:
        new_key=str(key.strip())+'_'+str(value.strip())
        return new_key.strip(), 1



def transform_T_F(dictionary):
    new_dict={}
    k , v = dictionary
    if v.strip()[0]!='{':
        new_key, new_value=key_value(k.strip(),v.strip())
        new_dict[new_key]=new_value
    else:
        for sub_k, sub_v in ast.literal_eval(v.strip()).items():
            new_key=k.strip()+'_'+sub_k.strip()
            new_key, new_value=key_value(new_key, str(sub_v))
            new_dict[new_key]=new_value
    return new_dict

def dict_merge(a, b):
    c = a.copy()
    c.update(b)
    return c

```

```python
for i, row in enumerate(df['attributes']):
    if row:
        row_dict={}
        for item in row:
            items_splitted=item.split(':', 1)
            items_in_dict=transform_T_F(items_splitted)
            row_dict = dict_merge(row_dict,items_in_dict)
        df.set_value(i, 'attributes_new', row_dict)
        
```
